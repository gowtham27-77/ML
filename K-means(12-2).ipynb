```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

Clustering - Divide the universities in to groups (Clusters)

```python
Univ = pd.read_csv("Universities.csv")
Univ
```

#### Standardization of the data

```python
#Read all numeric columns in to Univ1
Univ1 = Univ.iloc[:,1:]
Univ1
```

```python
cols = Univ1.columns
```

```python
# standardistion function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Univ_df = pd.DataFrame(scaler.fit_transform(Univ1),columns = cols)
scaled_Univ_df
```

```python
#build 3 clusters using Kmeans cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=0) 
clusters_new.fit(scaled_Univ_df)
```

```python
# print the cluster labels
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
#assign clusters to the Univ data set
Univ['clusterid_new'] = clusters_new.labels_
Univ
```

```python
Univ.sort_values(by = "clusterid_new")
```

```python
# Use groupby() to find aggregated (mean) values in each cluster
Univ.iloc[:,1:].groupby("clusterid_new").mean()
```

#### Observations
- cluster 2 appears to be the top rated universities cluster as the cut off score, Top10, SFRatio parameter mean values are highest
- cluster 1 appears to occupy the middle level rated universities
- cluter 0 comes as the lower level rated universities

```python
# optimal graph using k value
wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_Univ_df)
    wcss.append(kmeans.inertia_)
print(wcss)
plt.plot(range(1,20), wcss)
plt.title('Eblow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Quality  of clusters is expressed in terms
# of silhoutte score
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_Univ_df, clusters_new.labels_, metric='euclidean')
score
```
